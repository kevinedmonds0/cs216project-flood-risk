```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Cleaning Dataset

```python
#annual house price
dfOld = pd.read_csv("Data/Housing Price Index/Annual House Price Indexes/hpi_at_bdl_tract.csv", low_memory=False)
dfOld = dfOld.dropna()
dfHPI = dfOld.copy()
```

```python
#master file
dfOld = pd.read_csv("Data/Housing Price Index/Master Files/hpi_master.csv", low_memory=False)
dfOld = dfOld.dropna()
dfMaster = dfOld.copy()
```

```python
#metro
dfOld = pd.read_csv("Data/Housing Price Index/Quarterly Data/All Transactions Indexes /hpi_at_metro.csv", names=["metro_name", "metro_code", "year", "quarter", "hpi", "annual_change"], low_memory=False)
dfOld = dfOld.dropna()
dfMetro = dfOld.copy()
```

```python
#state
dfOld = pd.read_csv("Data/Housing Price Index/Quarterly Data/All Transactions Indexes /hpi_at_state.csv", names=["state_abbr", "year", "quarter", "hpi"], low_memory=False)
dfOld = dfOld.dropna()
dfState = dfOld.copy()
```

```python
#us and census
dfOld = pd.read_csv("Data/Housing Price Index/Quarterly Data/All Transactions Indexes /hpi_at_us_and_census.csv", low_memory=False)
dfOld = dfOld.dropna()
dfCensus = dfOld.copy()
```

### Initial Analysis

```python
dfMetro["hpi"] = pd.to_numeric(dfMetro["hpi"], errors="coerce")
dfState["hpi"] = pd.to_numeric(dfState["hpi"], errors="coerce")


metro_trend = dfMetro.groupby("year")["hpi"].mean()
state_trend = dfState.groupby("year")["hpi"].mean()

plt.figure(figsize=(10,5))

plt.plot(metro_trend, label="Metro Avg", linestyle="--")
plt.plot(state_trend, label="State Avg")

plt.title("Metro vs State HPI Comparison")
plt.xlabel("Year")
plt.ylabel("HPI")
plt.legend()
plt.show()
```
